```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo_number_13 = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
food_db = mongo_number_13['uk_food']
```

```python
# review the collections in our database
print(food_db.list_collection_names())
```

```python
# assign the collection to a variable
food_establishments = food_db['establishments']
print(food_db.establishments.find_one())
```

```python
# Find the establishments with a hygiene score of 20

#set query
hy_query = {'scores.Hygiene':{'$eq':20}}
#set query results
hy_results = food_establishments.find(hy_query)
#compute the count of documents
hy_count = food_establishments.count_documents(hy_query)
#perform verification that query is successful
hy_check = food_establishments.find_one(hy_query)

# Use count_documents to display the number of documents in the result
# Display the first document in the results using pprint

print(f"Number of documents with a hygiene equal to 20 is: {hy_count}\nWith the first result as the following: \n{hy_check}")
```

```python
# Convert the result to a Pandas DataFrame
hy_df = pd.DataFrame(hy_results)

# Display the number of rows in the DataFrame
print("Number of Rows in Hy DataFrame: ", len(hy_df))

# Display the first 10 rows of the DataFrame
hy_df.head(10)
```

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.

#set query
lar_query = {"$and":[{"AddressLine3":"London"},{"RatingValue":{"$gt":4}}]}
#set query results
lar_results = food_establishments.find(lar_query)
#compute the count of documents
lar_count = food_establishments.count_documents(lar_query)
#perform verification that query is successful
lar_check = food_establishments.find_one(lar_query)

# Use count_documents to display the number of documents in the result
# Display the first document in the results using pprint

print(f"Number of documents with local authority name equal to London and a rating value higher than 4 is: {lar_count}\nWith the first result as the following: \n{lar_check}")

# Use count_documents to display the number of documents in the result

# Display the first document in the results using pprint
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
#Matches establishments with a hygiene score of 0
hy_match = {'$match': {'scores.Hygiene': {'$eq': 0}}}

# 2. Groups the matches by Local Authority
hy_group = {'$group': {'_id': "$LocalAuthorityName", 'count': { '$sum': 1 }}}

# 3. Sorts the matches from highest to lowest
hy_sort = {'$sort': {'count': -1 }}

hy_pipeline = [hy_match,hy_group,hy_sort]

hy_lar_results = list(food_establishments.aggregate(hy_pipeline))

# Print the number of documents in the result
print("Number of establishments in the results: ", len(hy_lar_results))

# Print the first 10 results
pprint(hy_lar_results[0:10])
hy_pipeline
```

```python
# Convert the result to a Pandas DataFrame
hy_lar_df = pd.DataFrame(hy_lar_results)

# Display the number of rows in the DataFrame
print("Number of Rows in lar DataFrame: ", len(hy_lar_df))

# Display the first 10 rows of the DataFrame
hy_lar_df.head(5)
```
